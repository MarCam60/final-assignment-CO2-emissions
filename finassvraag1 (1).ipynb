```python
df_2 = df_1.loc[df_1['year'] > 2015]
#df_2.nlargest(10, ['co2_per_capita'])
df_2.nsmallest(10, ['co2_growth_prct'])
# bekijk de landen met de laagste groeipercentages
```

```python
import numpy as np

import math
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
correlaties = df_2.corr()
sns.heatmap(correlaties)
# zoek verbandne, zodat kan worden bepaald welke de beste voorspeller zijn van co2 uitstoot en dan met name per hoofd van de bevolking. Ik kies er bwoust voor om per hoofd van de bevolking te onderzoeken en niet per landalleen.
```

```python
df_corr = df_2.corr(method ='pearson')
df_corr
```

```python
df_kendall = df_corr.corr(method='kendall')
df_kendall
```

```python
from pylab import rcParams
import seaborn as sb
```

```python
#Now what does this mean??
#The dark shades indicate a strong negative corelation
#The lighter shades indicate a strong positive corelation
#i.e if any variable is plotted against itself, it'll have a corelation of 1
rcParams['figure.figsize'] = 14.7,8.27
sb.heatmap(df_kendall, 
           xticklabels=df_kendall.columns.values, 
           yticklabels=df_kendall.columns.values, 
           cmap="YlGnBu",
          annot=True)
```

Uit het bovenstaande valt duidelijk aan te geven welke bronnen het best de co2 uitstoot voorspellen . dat zijn ,gas, consumptie en energy steeds per capita. Ik heb per capita genomen omdat ik dan ook direct verwerk de bevolkingsmutatie.
