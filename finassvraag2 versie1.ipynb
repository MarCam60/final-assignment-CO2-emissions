```python
df_1.nlargest(10, ['co2_per_capita'])
```

```python
df_2 = df_1.loc[df_1['year'] > 2015]
#df_2.nlargest(10, ['co2_per_capita'])
df_2.nsmallest(10, ['co2_growth_prct'])
```

```python
df_2.columns
```

```python
df_2.drop (['gdp', 'cement_co2', 'cement_co2_per_capita', 'co2',
       'co2_growth_abs', 'co2_growth_prct',  'co2_per_gdp',
       'consumption_co2_per_capita', 'consumption_co2_per_gdp',
       'energy_per_capita', 'gas_co2_per_capita', 'oil_co2_per_capita',
       'trade_co2'], axis = 1, inplace = True)
```

```python
df_2
```

```python
df_2016 = df_2[df_2["year"] == 2016].rename(columns={"co2_per_capita": "2016"}).drop(columns="year")
df_2017 = df_2[df_2["year"] == 2017].rename(columns={"co2_per_capita": "2017"}).drop(columns="year")
df_2018 = df_2[df_2["year"] == 2018].rename(columns={"co2_per_capita": "2018"}).drop(columns="year")
```

```python
df_2016
```

```python
from numpy import outer
df_2016_2018= df_2016.merge(df_2018, on = 'country')
df_2016_2018.head()
# 2 jaren gekozen waar vergelijkingen mee gemaakt worden
```

```python
df_2016_2018.set_index('country', inplace =True)
df_2016_2018.head()
```

```python
df_2016_2018['prct_change'] = (df_2016_2018["2018"] / df_2016_2018["2016"]) * 100
df_2016_2018.head()
# de procentuele verandering berekenen
```

```python
df_co2_biggest = df_2016_2018.sort_values(by=['prct_change']).head(25)
df_co2_biggest
# filteren van de procentuele verandering. Landen met de kleinste procentuele verandering komen bovenaan te staan.Hoe kleiner hoe beter voor de conclusie. Uit de cijfers blijkt dat Colombia, Mozambique en Ivoorkust het beste scoren.
```

```python
import matplotlib.pyplot as plt
```

```python
df_plot_strides = df_co2_biggest.drop(columns=["2016", "2018"])
df_10 = df_plot_strides.head(10)
```

```python
df_10.plot(title = "co2 biggest steps per capita", xlabel= 'country', ylabel='percentage change per capita 2016/2018', kind = "bar", figsize = (25,10))
plt.legend()
```

```python
df_10.plot(title = "co2 biggest steps per capita", xlabel= 'country', ylabel='percentage change per capita 2016/2018', kind = "pie", figsize = (25,10), subplots = True)
plt.legend()
```
